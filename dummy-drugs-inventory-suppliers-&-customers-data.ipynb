```python
import pandas as pd
import string
import random
```

```python
# Generate a random 6-digit PIN code
def generate_pin():
    return ''.join(random.choice(string.digits) for _ in range(6))

# Generate a random 10-digit phone number
def generate_phone():
    return ''.join(random.choice(string.digits) for _ in range(10))

# Generate a random email address
def generate_email(company_name):
    domains = ['gmail.com', 'yahoo.com', 'outlook.com', 'example.com']
    domain = random.choice(domains)
    email = f"{company_name.lower().replace(' ', '_')}@{domain}"
    return email

# Generate random supplier data
def generate_supplier_data(id):
    company_name = f"Supplier {id}"
    address_1 = f"{random.choice(string.ascii_uppercase)}{random.choice(string.ascii_lowercase)}{random.choice(string.ascii_lowercase)} {random.choice(['Street', 'Avenue', 'Road'])}"
    address_2 = f"{random.choice(['Mumbai', 'Delhi', 'Chennai', 'Kolkata', 'Bangalore'])}"
    pin = generate_pin()
    phone = generate_phone()
    email = generate_email(company_name)

    return {'id': id, 'name': company_name, 'address_1': address_1, 'address_2': address_2, 'pin': pin, 'phone': phone, 'email': email}

# Generate 20 random supplier company names
supplier_data = [generate_supplier_data(id) for id in range(1, 21)]

# Print the generated supplier data
for supplier in supplier_data:
    print(supplier)
```

```python
import random
import string

# Generate a random 6-digit PIN code
def generate_pin():
    return ''.join(random.choice(string.digits) for _ in range(6))

# Generate a random 10-digit phone number
def generate_phone():
    return ''.join(random.choice(string.digits) for _ in range(10))

# Generate a random email address
def generate_email(shop_name):
    domains = ['gmail.com', 'yahoo.com', 'outlook.com', 'example.com']
    domain = random.choice(domains)
    email = f"{shop_name.lower().replace(' ', '_')}@{domain}"
    return email

# Generate random drug shop data
def generate_drug_shop_data(id):
    shop_name = f"Drug Store {id}"
    address_1 = f"{random.choice(string.ascii_uppercase)}{random.choice(string.ascii_lowercase)}{random.choice(string.ascii_lowercase)} {random.choice(['Street', 'Avenue', 'Road'])}"
    address_2 = f"{random.choice(['Mumbai', 'Delhi', 'Chennai', 'Kolkata', 'Bangalore'])}"
    pin = generate_pin()
    phone = generate_phone()
    email = generate_email(shop_name)
    owner = f"{random.choice(['Amit', 'Rajesh', 'Deepak', 'Neha', 'Priya'])}"

    return {'id': id, 'name': shop_name, 'address_1': address_1, 'address_2': address_2, 'pin': pin, 'phone': phone, 'email': email, 'owner': owner}

# Generate 30 random drug store names
drug_shop_data = [generate_drug_shop_data(id) for id in range(1, 31)]

# Print the generated drug shop data
for shop in drug_shop_data:
    print(shop)
```
