```python
import pandas as pd
import random
import string
```

```python
# Generate a random 12-character password
def generate_password():
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for _ in range(12))
    return password

# Generate a random email address
def generate_email(username):
    domains = ['gmail.com', 'yahoo.com', 'hotmail.com', 'example.com']
    domain = random.choice(domains)
    email = f"{username}@{domain}"
    return email

# Generate random user data
def generate_user_data():
    id = random.randint(1, 1000)
    
    # Generate a random first name and last name
    first_names = ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack']
    last_names = ['Smith', 'Johnson', 'Brown', 'Davis', 'Wilson', 'Clark', 'Hall', 'Lee', 'Moore', 'Adams']
    name = f"{random.choice(first_names)} {random.choice(last_names)}"
    
    password = generate_password()
    username = name.lower().replace(' ', '_')
    email = generate_email(username)

    return {'id': id, 'name': name, 'password': password, 'email': email}

# Generate 10 random user names
user_data = [generate_user_data() for _ in range(10)]

# Print the generated user data
for user in user_data:
    print(user)
```
