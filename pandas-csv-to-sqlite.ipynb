```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('meds.db') 
  
# Load CSV data into Pandas DataFrame 
meds_data = pd.read_csv('drugs-inventory-products_medicines-data.csv') 
# Write the data to a sqlite table 
meds_data.to_sql('meds', conn, if_exists='replace', index=False) 
  
# Create a cursor object 
cur = conn.cursor() 
# Fetch and display result 
for row in cur.execute('SELECT * FROM meds'): 
    print(row) 
# Close connection to SQLite database 
conn.close() 
```
